```python
metadata_augmented_path = Path("../data/augmented/metadata.csv")
df_augmented = pd.read_csv(metadata_augmented_path)
```

```python
df_filtered = df_augmented[
    df_augmented["image_path"].str.contains("_color|_spatial", regex=True)
]
```

```python
nb_total = df_augmented["image_path"].count()

nb_color = df_filtered["image_path"].str.contains("_color", regex=False).sum()
nb_spatial = df_filtered["image_path"].str.contains("_spatial", regex=False).sum()

print("Total images   :", nb_total)
print("Images color   :", nb_color)
print("Images spatial :", nb_spatial)
```
